```python
########  main code #############

import cv2
import json
import os

with open("FLIR_ADAS_v2/images_thermal_val/index.json", "r") as file:
    ground_truth = json.load(file)
    
# Set the path to the directory containing the images
path = "FLIR_ADAS_v2/images_thermal_val/data"
bounding_boxes = []

# Iterate through all the files in the directory
for filename in os.listdir(path):
    # Check if the file is an image
    if filename.endswith(".jpg") or filename.endswith(".png"):
        imgID = filename.split("-")[-1].split(".")[0]
        # Read the image using cv2.imread()
        gray = cv2.imread(os.path.join(path, filename))
        
        for i, frames in enumerate(ground_truth["frames"]):
            if ground_truth["frames"][i]["datasetFrameId"] == imgID:
                index = i
        for j, annotation in enumerate( ground_truth["frames"][index]['annotations']):
            bounding_box = annotation['boundingBox']
            label = annotation['labels']
            print("for bounding Box:", j+1," ",label[0])
            x, y, w, h = bounding_box['x'],bounding_box['y'],bounding_box['w'],bounding_box['h']
            cv2.rectangle(gray, (x, y), (x + w, y + h), (0, 255, 0), 1)
            cv2.putText(gray, str(label[0]), (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1)

        cv2.imshow("Thermal Image with Ground Truth", gray)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    print("-----------------------------------------------------------------------------")   
```
