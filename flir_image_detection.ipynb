```python
import os
import cv2
import matplotlib.pyplot as plt
from utils import *
from darknet import Darknet
```

```python
cfg_file = 'yolov3.cfg'
weight_file = 'yolov3.weights'
namesfile = 'coco.names'
m = Darknet(cfg_file)
m.load_weights(weight_file)
class_names = load_class_names(namesfile)

plt.rcParams['figure.figsize'] = [24.0, 14.0]
data_dir = "FLIR_ADAS_v2/images_rgb_val/data"
iou_thresh = 0.4
nms_thresh = 0.7
```

```python

for im in os.listdir(data_dir):
    img = cv2.imread(os.path.join(data_dir, im))    
    re_im = cv2.resize(img, (m.width, m.height))
    boundaries = detect_objects(m, re_im, iou_thresh, nms_thresh)
    plot_boxes(img, boundaries, class_names, plot_labels = True)
```
